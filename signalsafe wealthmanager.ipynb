```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Parameters
sample_rate = 1000  # Hz (samples per second)
duration = 1.0      # seconds
frequency = 10.0    # Hz (change this to the desired brain wave frequency)

# Create a time tensor
t = torch.arange(0, duration, 1/sample_rate)

# Create a sinusoidal waveform
waveform = torch.sin(2 * np.pi * frequency * t)

# Convert tensor to numpy array for plotting
t_np = t.numpy()
waveform_np = waveform.numpy()

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(t_np, waveform_np)
plt.title(f'Sinusoidal Waveform at {frequency} Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Parameters
sample_rate = 1000  # Hz (samples per second)
duration = 5.0      # seconds
frequencies = [4, 8, 12, 30]  # Delta, Theta, Alpha, Beta

# Create a time tensor
t = torch.arange(0, duration, 1/sample_rate)

# Generate a composite waveform
waveform = torch.zeros_like(t)
for freq in frequencies:
    waveform += torch.sin(2 * np.pi * freq * t)

# Normalize the waveform
waveform /= len(frequencies)

# Convert tensor to numpy array for plotting
t_np = t.numpy()
waveform_np = waveform.numpy()

# Plot the waveform
plt.figure(figsize=(12, 6))
plt.plot(t_np, waveform_np)
plt.title('Composite Sinusoidal Waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Parameters
sample_rate = 1000  # Hz (samples per second)
duration = 5.0      # seconds
initial_freq = 4    # Hz (starting frequency)
final_freq = 30     # Hz (ending frequency)

# Create a time tensor
t = torch.arange(0, duration, 1/sample_rate)

# Frequency modulation: linearly increase frequency from initial to final
frequency_modulation = initial_freq + (final_freq - initial_freq) * (t / duration)

# Create a sinusoidal waveform with modulated frequency
waveform = torch.sin(2 * np.pi * frequency_modulation * t)

# Convert tensor to numpy array for plotting
t_np = t.numpy()
waveform_np = waveform.numpy()

# Plot the waveform
plt.figure(figsize=(12, 6))
plt.plot(t_np, waveform_np)
plt.title('Frequency Modulated Sinusoidal Waveform')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
